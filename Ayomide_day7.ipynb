```python
import pandas as pd
import numpy as np
```

```python
air = pd.read_csv("AirQualityUCI.csv",sep=';')
```

```python
air=air.dropna(axis=1,how='all')
air=air.dropna(axis=0,how='any')
```

```python
air.head()
```

```python
#identify the month
dates=list(air['Date'])
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_feature = []
for i in dates:
    if type(i)==str:
        split_i = i.split('/')
        month = int(split_i[1])-1
        #print(month)
        month_feature += [months[month]]
    else:
        month_feature += [None]
air['Months']=month_feature
```

```python
#Morning,Afternoon,Evening
times=list(air['Time'])
time_feature = []
for i in times:
    if type(i)==str:
        split_i = i.split('.')
        time = int(split_i[0])
        #print(month)
        if time < 8:
            time_feature += ['Morning']
        elif time >= 8 and time< 16:
            time_feature += ['Afternoon']
        elif time >= 16 and time<24:
            time_feature += ['Evening']
       
    else:
        time_feature += [None]
air['Morn/After/Even']=time_feature
```

```python
#identify weekdays with True
dates=list(air['Date'])
weekends = ['sunday','saturday']
week_feature = []
for i in dates:
    if type(i)==str:
        i_split = i.split('/')
        i_split.reverse()
        #print(i_split)
        new_date = '-'.join(i_split)
        #print(new_date)
        day = pd.Timestamp(str(new_date)).day_name()
        #print(day)
        if day.lower() in weekends:
            week_feature += ['False']
        elif day.lower() not in weekends:
            week_feature += ['True']
    else:
        week_feature += [None]
air["Weekdays(T/F)"]=week_feature
```
