```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
df = pd.DataFrame(wine.data,columns=wine.feature_names)
df["target"]=wine.target
```

```python
df.head()
```

```python
df_normalize = (df - df.min())/(df.max()-df.min()) #mormalize using minmax method
```

```python
df_normalize.head()
```

```python
df.corr(method = "kendall")  #kendall
```

```python
df.corr(method = "pearson") #pearson
```

```python
pearson_corr= df.corr(method = "pearson") #heatmap
plt.pcolor(pearson_corr)
plt.show()
```

```python
pearson_corr.target.abs().sort_values()
```
