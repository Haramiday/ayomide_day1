```python
new_cam = cam.dropna(axis=0,how='any')
```

```python
spec=list(new_cam['specialisation'])
```

```python
specElement = {}
for i in spec:
    if i not in specElement:
        specElement[i]=0
    else:
        specElement[i] += 1
key = list(specElement.keys())

f_key=key[0]
for k in range(1,len(key)):
    if specElement[f_key]>specElement[key[k]]:
        f_key = f_key
    else:
        f_key = key[k]
        
f_key #this is the demanded specialisation
```
